```python
# menambahkan pustaka yang diperlukan
from sklearn import datasets
from sklearn import model_selection
from sklearn.metrics import accuracy_score
import numpy as np
import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

#array untuk menampilkan hasil klasifikasi
flower = ['Iris-Setosa', 'Iris-versiColor', 'Iris-Virginica']

# mengambil sample dataset bunga iris yang disediakan tensorflow 
iris=datasets.load_iris()
X = iris.data
Y = iris.target
y = tensorflow.keras.utils.to_categorical(Y)

# proses ini membagi dataset menjadi 2, data training dan dataset...
# test_size = menyatakan seberapa porsi untuk test dan train
crossValidation = model_selection.KFold(n_splits=10, random_state=14, shuffle=True)

# cetak sample fitur data iris 
print("input")
print(X)
# cetak klasifikasi fitur
print("output")
print(Y)
```

```python
# membuat model Neural Network
model = Sequential()
# membuat layer input dengan 4 masukan data fitur, 8 neuron, dan fungsi aktivasi relu
# sepal length, sepal width, petal length, petal width
model.add(Dense(8, input_dim=4, activation='relu'))
# membuat layer hidden dengan 8 neuron
model.add(Dense(8, activation='relu'))
# membuat layer output dengan 3 output klasifikasi (setosa, versicolor, virginica), dan fungsi aktivasi softmax
model.add(Dense(3, activation='softmax'))
# melakukan kompilasi model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
# melakukan proses training data sample
model.fit(X,y, batch_size=5, epochs=50, verbose=2)
# melakukan penyimpanan hasil training
model.save('iris.h5')
# mencetak desain model Neural Network
print(model.summary())
# mendapatkan nilai bobot
print(model.get_weights())
```

```python
# melakukan prediksi pada sample data tes Iris-Setosa
dataTes = [[5.1, 3.5, 1.4, 0.2]]
prediction = model.predict(dataTes)
print(prediction)
pos_indext_nilai_max = np.array(prediction).argmax()
print(pos_indext_nilai_max)
print("Bunga ini diklasifikasikan sebagai : ")
print(flower[pos_indext_nilai_max])
```
